```python
import os
import pandas as pd

# Define the root directory where the files are stored
root_directory = "/content/training-datalab-gold-standard-toxicity-8923a14/data/"

# List to store extracted data
data = []

# Traverse through the directory
for parent_folder in os.listdir(root_directory):
    parent_path = os.path.join(root_directory, parent_folder)

    if os.path.isdir(parent_path):  # Check if it's a folder
        for sub_folder in os.listdir(parent_path):
            sub_path = os.path.join(parent_path, sub_folder)

            if os.path.isdir(sub_path):  # Check if it's a subfolder
                for file in os.listdir(sub_path):
                    file_path = os.path.join(sub_path, file)

                    if file.endswith(".txt"):  # Process only text files
                        with open(file_path, "r", encoding="utf-8") as f:
                            content = f.read().strip()  # Read the file content

                        # Append extracted data
                        data.append([parent_folder, sub_folder, file, content])

# Create DataFrame
df = pd.DataFrame(data, columns=["Parent Folder", "Subfolder", "Filename", "Content"])

# Save to CSV
csv_path = "extracted_data.csv"
df.to_csv(csv_path, index=False)

df
```
